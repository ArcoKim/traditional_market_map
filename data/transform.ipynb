```python
import pandas as pd
```

```python
market = pd.read_csv('소상공인시장진흥공단_전통시장_20240722.csv')
market.head()
```

```python
market = market.drop(columns=['사용가능상품권', '주차장보유여부', '데이터기준일자'])
market.head()
```

```python
market.columns = ["name", "category", "address1", "address2", "open_cycle", "latitude", "longitude", "number_of_stores", 
                  "items_type", "homepage", "restroom", "open_year", "phone_number"]
market.info()
```

```python
market = market.astype({'number_of_stores': 'Int64', 'open_year': 'Int64'})
market[['number_of_stores', 'open_year']]
```

```python
market[market['address1'].isnull()]
```

```python
mv = market['name'].value_counts()
mv[mv > 1]
```

```python
idx = mv[mv > 1].index.tolist()
idx
```

```python
def get_name(row):
    if row['name'] in idx:
        zone = row['address1'].split()[1]
        if len(zone) == 2:
            return zone + row['name']
        return zone[:-1] + row['name']
    return row['name']

market['join_name'] = market.apply(get_name, axis=1)
market['join_name']
```

	id INTEGER PRIMARY KEY AUTOINCREMENT,

	elevator BOOLEAN NOT NULL,
	support_center BOOLEAN NOT NULL,

	gym BOOLEAN NOT NULL,
	library BOOLEAN NOT NULL,

	cooperative_warehouse BOOLEAN NOT NULL,
	parking BOOLEAN NOT NULL, 
	education_center BOOLEAN NOT NULL,
	meeting_room BOOLEAN NOT NULL,
	AED BOOLEAN NOT NULL

CREATE TABLE market (
	id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    category TEXT NOT NULL,
    address TEXT NOT NULL,
    city1 TEXT NOT NULL,
    city2 TEXT NOT NULL,
    open_cycle TEXT NOT NULL,
    latitude FLOAT NOT NULL,
	longitude FLOAT NOT NULL,
	number_of_stores INT, 
	items_type TEXT NOT NULL,
	homepage TEXT,
	open_year INT,
	phone_number TEXT,
    facility_id INT NOT NULL UNIQUE,
    FOREIGN KEY (facility_id) REFERENCES facility(id)
);
