```python
import pandas as pd
```

```python
market = pd.read_csv('소상공인시장진흥공단_전통시장_20240722.csv')
market.head()
```

```python
market = market.drop(columns=['사용가능상품권', '주차장보유여부', '데이터기준일자'])
market.head()
```

```python
market.columns = ["name", "category", "address1", "address2", "open_cycle", "latitude", "longitude", "number_of_stores", 
                  "items_type", "homepage", "restroom", "open_year", "phone_number"]
market.info()
```

```python
market = market.astype({'number_of_stores': 'Int64', 'open_year': 'Int64'})
market[['number_of_stores', 'open_year']]
```

```python
facility = pd.read_csv('소상공인시장진흥공단_전통시장현황_20240719.csv', encoding='cp949')
facility.head()
```

```python
facility = facility.drop(columns=['시장코드', '시장 유형', '지번주소'])
facility.head()
```

```python
facility.columns = ["name", "address1", "city1", "city2", "arcade", "elevator", "support_center", "sprinkler", "fire_detector", 
                    "play_room", "call_center", "foyer", "lactation_center", "stockroom", "bicycle_rack", "gym", "library",
                    "shopping_cart", "visitor_center", "movement_path", "broadcasting_center", "cultural_center", 
                    "cooperative_warehouse", "parking", "education_center", "meeting_room", "AED"]
facility.info()
```

```python
alls = pd.merge(market, facility, on=["name", "address1"])
alls.info()
```

```python
market = alls[["name", "category", "address1", "address2", "city1", "city2", "open_cycle", "latitude", "longitude", "number_of_stores",
               "items_type", "homepage", "open_year", "phone_number"]]

market.index += 1
market = market.reset_index().rename(columns={"index": "facility_id"})

market.index += 1
market = market.reset_index().rename(columns={"index": "id"})

market.head()
```
